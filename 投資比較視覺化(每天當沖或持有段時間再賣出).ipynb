# 試算出股價平均值

```python
stock = [321,200,366,211,223,221,555,112,123,654]
```

```python
type(stock)
```

## 方法一

```python
sum1=0

for i in range(len(stock)):
    sum1=sum1+stock[i]
print(sum1/len(stock))
```

## 方法二

```python
sum1=0

for i in stock:
    sum1=sum1+i
print(sum1/len(stock))
```

# 算出總額

```python
stock_name=["apple","IBM","HP","BMW","SAMSONG"]
stock = [321,200,366,211,223]
stock_num = [1,2,3,4,5]
list3=[]
total=0

for i in range(len(stock)):
    list3.append(stock[i]*stock_num[i])
print(list3)

for j in list3:
    total=total+j
print(total)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
plt.plot(stock,"o--")
```

```python
plt.figure(figsize=(8,6))
plt.pie(list3,labels=stock_name,autopct='%1.1f%%')
```
