# 運用pandas讀取excel

```python
import pandas as pd
```

```python
stock_0050 = pd.read_csv('0050_07.csv',encoding='cp950',skiprows=1)
stock_2330 = pd.read_csv('2330_07.csv',encoding='cp950',skiprows=1)
```

# 將不需要col及row拿掉(這裡是只取要的row)

```python
stock_0050 = stock_0050.iloc[0:10,:]
stock_2330 = stock_2330.iloc[0:10,:]
```

```python
stock_0050
```

# 將空白欄拿掉 

```python
stock_0050.dropna(axis=1,inplace=True)
stock_2330.dropna(axis=1,inplace=True)
```

```python
stock_0050
```

# 將成交股數的","取代，並轉為int型態

```python
stock_0050['成交股數'] = stock_0050['成交股數'].str.replace(',','').astype(int)
stock_2330['成交股數'] = stock_2330['成交股數'].str.replace(',','').astype(int)
```

```python
stock_0050
```

# 畫圖!

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.figure(figsize=(15,8))
plt.bar(stock_0050['日期'],stock_0050["成交股數"])
plt.title("成交股數",fontsize=25)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('日期',fontsize=15)
plt.ylabel('當月成交股數',fontsize=15)
plt.legend()
plt.grid()
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.figure(figsize=(15,8))
plt.bar(stock_2330['日期'],stock_2330["成交股數"])
plt.title("成交股數",fontsize=25)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('日期',fontsize=15)
plt.ylabel('當月成交股數',fontsize=15)
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(15,6))
plt.plot(stock_2330['日期'],stock_2330["成交股數"],"o-",label="2330")
plt.plot(stock_0050['日期'],stock_0050["成交股數"],"o-",label="0050")
plt.title("成交股數",fontsize=25)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('日期',fontsize=15)
plt.ylabel('當月成交股數',fontsize=15)
plt.legend()
plt.grid()
```
