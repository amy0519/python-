```python
si = np.array(df.loc[df['非擔任主管職務之全時員工資訊-員工薪資-平均數(仟元/人)-108年']>700,'非擔任主管職務之全時員工資訊-員工薪資-平均數(仟元/人)-108年'])
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.figure(figsize=(15,10))
plt.scatter(df[df['非擔任主管職務之全時員工資訊-員工薪資-平均數(仟元/人)-108年']>700].公司名稱,df.loc[df['非擔任主管職務之全時員工資訊-員工薪資-平均數(仟元/人)-108年']>700,'非擔任主管職務之全時員工資訊-員工薪資-平均數(仟元/人)-108年']/10,c=np.log2(si),s=0.0001*si**2.3)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
plt.title('108年半導體上市公司平均員工薪資費用',fontsize=25)
```
