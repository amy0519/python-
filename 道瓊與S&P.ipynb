```python
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import datetime
```

```python
start = datetime.datetime(2020,1,1)
end = datetime.datetime(2020,7,1)
df_stock = pdr.DataReader('2330.tw','yahoo',start,end)
```

```python
df_stock
```

```python
stock_number = ['DJIA','^GSPC']
start = datetime.datetime(2020,1,1)
end = datetime.datetime(2020,7,1)
df_stock = pdr.DataReader(stock_number,'yahoo',start,end)
```

```python
df_stock.Close[['DJIA']].tail()
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.figure(figsize=(15,8))
plt.plot(df_stock.Close['DJIA'].rolling(15).mean(),".-r",label="DJIA mean")
plt.plot(df_stock.Close['DJIA'],"o-",label="DJIA")
plt.title("成交股數",fontsize=25)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('time',fontsize=15)
plt.ylabel('Dow Jones',fontsize=15)
plt.legend()
plt.grid()
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.figure(figsize=(15,8))
plt.plot(df_stock.Close['^GSPC'].rolling(15).mean(),".-r",label="^GSPC mean")
plt.plot(df_stock.Close['^GSPC'],"o-",label="^GSPC")
plt.title("成交股數",fontsize=25)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('time',fontsize=15)
plt.ylabel('Dow Jones',fontsize=15)
plt.legend()
plt.grid()
```

```python
fig, ax1 = plt.subplots(figsize=(20,15))
ax1.plot(df_stock.Close.DJIA.rolling(15).mean(),'.-r',label="DJIA mean")
ax1.plot(df_stock.Close.DJIA,'.-g')
ax1.set_xlabel('time(year)',fontsize=25)
ax1.set_ylabel('Dow Jones',fontsize=25)
ax1.legend(loc=0)

ax2 = ax1.twinx()  #ax1,ax2 公用y軸
ax2.plot(df_stock.Close['^GSPC'].rolling(15).mean(),'.-b',label="^GSPC mean")
ax2.set_ylabel('S&P 500',fontsize=25)
ax2.plot(df_stock.Close['^GSPC'],'.-y')
ax2.legend(loc=1)
ax1.grid()

plt.xticks(fontsize=25)
plt.yticks(fontsize=20)
```
